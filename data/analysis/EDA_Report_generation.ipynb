# Importing requried modules

```python
import os
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
from google.colab import drive
```

# Generating report for train.csv

```python
drive_path = "/content/drive/MyDrive"
dataset_path = f"{drive_path}/Datasets"
aux_dataset_path = f"{drive_path}/Datasets/auxiliary-data"
eda_reports_path = f"{drive_path}/EDA Reports"
```

```python
train_df = pd.read_csv(f"{dataset_path}/train.csv")
train_profile = ProfileReport(train_df)
train_profile.to_file(f"{eda_reports_path}/train_report.html")
```

# Generating reports for auxillary data

```python
auxilary_dataset_file_names = os.listdir(f"{aux_dataset_path}")
auxilary_dataset_file_names
```

```python
for auxilary_file in auxilary_dataset_file_names:
    file_path = os.path.join(f"{aux_dataset_path}/{auxilary_file}")  # Create the full file path
    df = pd.read_csv(file_path)
    profile = ProfileReport(df, title=f'Pandas Profiling Report for {auxilary_file}', explorative=True)
    file_name = f"{eda_reports_path}/report-{auxilary_file}.html"
    profile.to_file(file_name)  # Save the report to an HTML file
```
